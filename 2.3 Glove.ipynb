```python
from glove import Corpus, Glove

corpus = [
    "I love machine learning",
    "Machine learning is amazing",
    "Natural language processing is a branch of AI"
]


tokenized_corpus = [sentence.lower().split() for sentence in corpus]

glove_corpus = Corpus()
glove_corpus.fit(tokenized_corpus, window=5)

glove_model = Glove(no_components=50, learning_rate=0.05)
glove_model.fit(glove_corpus.matrix, epochs=30, no_threads=4, verbose=True)
glove_model.add_dictionary(glove_corpus.dictionary)

print("Vector for 'machine':", glove_model.word_vectors[glove_model.dictionary['machine']])
```
